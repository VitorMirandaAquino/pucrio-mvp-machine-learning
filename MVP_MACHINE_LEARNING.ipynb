# <a id='descricao'>Descri√ß√£o do Problema</a>

O conjunto de dados **HR Analytics** √© um conjunto de dados multivariado que cont√©m informa√ß√µes sobre funcion√°rios de uma empresa, incluindo caracter√≠sticas demogr√°ficas, profissionais e comportamentais. O problema visa identificar fatores que contribuem para o desligamento de funcion√°rios.

## Hip√≥teses do Problema

As hip√≥teses em rela√ß√£o ao **comportamento dos funcion√°rios**:
- O n√≠vel de satisfa√ß√£o geral est√° baixo o que pode levar ao desligamento
- Os funcion√°rios est√£o precisando viajar muito
- Os funcion√°rios est√£o precisando fazer muitas horas extras

- Os funcion√°rios est√£o levando muito anos at√© serem promovidos

As hip√≥teses que levam ao **desligamento da empresa**:
- Os funcion√°rios mais novos est√£o se desligando mais da empresa por um comportamento da nova gera√ß√£o

```python
def plot_comparative_boxplots(df_results, group_by='modelo', figsize=(14, 6)):
    """
    Plota dois boxplots comparativos na mesma figura: Acur√°cia e F1-Score
    
    Par√¢metros:
    - df_results: DataFrame com os resultados dos experimentos
    - group_by: Vari√°vel para agrupar no eixo y ('balanceamento', 'scaler', 'modelo')
    - figsize: Tamanho da figura (largura, altura)
    """
    
    # Configura√ß√£o da figura com 2 subplots
    fig, axes = plt.subplots(1, 2, figsize=figsize)
    
    # Configura√ß√£o do estilo
    plt.style.use('default')
    
    # Boxplot da Acur√°cia
    sns.boxplot(data=df_results, y=group_by, x='acc_mean', hue=group_by, ax=axes[0], palette='viridis', legend=False)
    axes[0].set_title('Distribui√ß√£o da Acur√°cia por ' + group_by.capitalize(), fontsize=14, fontweight='bold')
    axes[0].set_xlabel('Acur√°cia M√©dia', fontsize=12)
    axes[0].set_ylabel(group_by.capitalize(), fontsize=12)
    
    # Boxplot do F1-Score
    sns.boxplot(data=df_results, y=group_by, x='f1_mean', hue=group_by, ax=axes[1], palette='plasma', legend=False)
    axes[1].set_title('Distribui√ß√£o do F1-Score por ' + group_by.capitalize(), fontsize=14, fontweight='bold')
    axes[1].set_xlabel('F1-Score M√©dio', fontsize=12)
    axes[1].set_ylabel(group_by.capitalize(), fontsize=12)
    
    # Ajustar layout
    plt.tight_layout()
    
    # Mostrar o gr√°fico
    plt.show()
```

```python
def pipeline_models_combination(X_train, y_train, models, balancing, scalers, n_splits=5, random_state=SEED):
    """
    Executa grid search sistem√°tico testando todas as combina√ß√µes de modelos, 
    t√©cnicas de balanceamento e m√©todos de escalonamento.
    
    Par√¢metros:
    -----------
    X_train : pd.DataFrame ou array-like
        Features de treino
    y_train : pd.Series ou array-like  
        Target de treino
    models : dict
        Dicion√°rio com modelos {'nome': modelo_instancia}
    balancing : dict
        Dicion√°rio com t√©cnicas de balanceamento {'nome': tecnica_instancia}
    scalers : dict
        Dicion√°rio com escalonadores {'nome': scaler_instancia}
    n_splits : int, default=5
        N√∫mero de folds para valida√ß√£o cruzada
    random_state : int, default=SEED
        Seed para reprodutibilidade
        
    Retorna:
    --------
    pd.DataFrame
        DataFrame com resultados ordenados por F1-Score decrescente, contendo:
        - balanceamento, scaler, modelo: combina√ß√£o testada
        - acc_mean, acc_std: acur√°cia m√©dia e desvio padr√£o
        - f1_mean, f1_std: F1-Score macro m√©dio e desvio padr√£o
        - precision_mean, precision_std: precis√£o macro m√©dia e desvio padr√£o
        - recall_mean, recall_std: recall macro m√©dio e desvio padr√£o
        
    Notas:
    ------
    - Usa make_scorer com zero_division=0 para evitar warnings de precis√£o/recall indefinidos
    
    Exemplo:
    --------
    >>> resultados = pipeline_combinacoes_modelos(X_train, y_train, models, balancing, scalers)
    >>> print(resultados.head())
    """
    
    results = []
    cv = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=random_state)

    # Loop triplo para todas as combina√ß√µes
    for bal_name, balancer in balancing.items():
        for scaler_name, scaler in scalers.items():
            for model_name, model in models.items():

                # Pipeline din√¢mico baseado na presen√ßa de balanceamento
                if balancer:
                    pipe = ImbPipeline(steps=[
                        ("scaler", scaler),
                        ("bal", balancer),
                        ("model", model)
                    ])
                else:
                    pipe = Pipeline(steps=[
                        ("scaler", scaler),
                        ("model", model)
                    ])

                # Valida√ß√£o cruzada com m√∫ltiplas m√©tricas
                acc_scores = cross_val_score(pipe, X_train, y_train, cv=cv, scoring="accuracy")
                f1_scores = cross_val_score(pipe, X_train, y_train, cv=cv, scoring="f1_macro")
                
                # Usando make_scorer para controlar o zero_division (evita warnings)
                precision_scorer = make_scorer(precision_score, average='macro', zero_division=0)
                recall_scorer = make_scorer(recall_score, average='macro', zero_division=0)
                
                precision_scores = cross_val_score(pipe, X_train, y_train, cv=cv, scoring=precision_scorer)
                recall_scores = cross_val_score(pipe, X_train, y_train, cv=cv, scoring=recall_scorer)

                results.append({
                    "balanceamento": bal_name,
                    "scaler": scaler_name,
                    "modelo": model_name,
                    "acc_mean": np.mean(acc_scores),
                    "acc_std": np.std(acc_scores),
                    "f1_mean": np.mean(f1_scores),
                    "f1_std": np.std(f1_scores),
                    "precision_mean": np.mean(precision_scores),
                    "precision_std": np.std(precision_scores),
                    "recall_mean": np.mean(recall_scores),
                    "recall_std": np.std(recall_scores)
                })

    # Retorna DataFrame ordenado por F1-Score
    df_results = pd.DataFrame(results)
    return df_results.sort_values(by="f1_mean", ascending=False)
    
```

```python
def calculate_metrics(y_true, y_pred, model_name):
    """
    Calcula m√©tricas de classifica√ß√£o (macro) e retorna em formato de dicion√°rio.

    Par√¢metros
    ----------
    y_true : array-like
        R√≥tulos verdadeiros.
    y_pred : array-like
        R√≥tulos preditos.
    model_name : str
        Nome do modelo avaliado.

    Retorna
    -------
    dict
        Dicion√°rio com: nome do modelo, acur√°cia, F1-score, precis√£o e recall (macro).
    """
    return {
        'Modelo': model_name,
        'Acur√°cia': accuracy_score(y_true, y_pred),
        'F1-Score': f1_score(y_true, y_pred, average='macro'),
        'Precis√£o': precision_score(y_true, y_pred, average='macro', zero_division=0),
        'Recall': recall_score(y_true, y_pred, average='macro', zero_division=0)
    }
```

```python
def grid_optimization_parameters(model, pipeline, parameters, X_train, y_train, random_state=SEED):
    """
    Executa a otimiza√ß√£o de hiperpar√¢metros utilizando GridSearchCV com valida√ß√£o cruzada estratificada,
    avalia o melhor modelo encontrado e verifica poss√≠veis sinais de overfitting.

    Par√¢metros
    ----------
    model : str
        Nome do modelo (usado apenas para logging/prints).
    pipeline : sklearn.pipeline.Pipeline
        Pipeline de pr√©-processamento + modelo a ser otimizado.
    parameters : dict
        Grade de hiperpar√¢metros a ser explorada pelo GridSearchCV.
    X_train : array-like ou pandas.DataFrame
        Conjunto de features de treino.
    y_train : array-like ou pandas.Series
        Conjunto de r√≥tulos de treino.
    random_state : int
        Semente para reprodutibilidade na valida√ß√£o cruzada.

    Retorna
    -------
    best_model : sklearn estimator
        Estimador treinado com os melhores hiperpar√¢metros encontrados.
    grid_search : GridSearchCV
        Objeto do GridSearchCV com todos os resultados da busca.

    Notas
    -----
    - A fun√ß√£o imprime os melhores hiperpar√¢metros, o F1-score m√©dio de valida√ß√£o 
      (via cross-validation) e o F1-score no treino com os melhores par√¢metros.
    - Tamb√©m calcula o "gap" entre treino e valida√ß√£o para indicar poss√≠veis sinais de overfitting.
    - A m√©trica utilizada na otimiza√ß√£o √© o F1-macro, mas o c√°lculo do gap pode ser adaptado 
      para outras m√©tricas, se necess√°rio.
    """
    print(f"üìä Iniciando otimiza√ß√£o de hiperpar√¢metros - {model}")
    print("=" * 60)

    # GridSearchCV
    grid_search = GridSearchCV(
        pipeline,
        parameters,
        cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=random_state),
        scoring='f1_macro',
        n_jobs=-1,
        verbose=1
    )

    # Treinamento
    grid_search.fit(X_train, y_train)

    # Resultados
    print(f"\n‚úÖ Melhores hiperpar√¢metros {model}: {grid_search.best_params_}")
    print(f"üéØ Melhor F1-Score (valida√ß√£o - crossval): {grid_search.best_score_:.4f}")

    # Modelo otimizado
    best_model = grid_search.best_estimator_

    # Score treino com os melhores par√¢metros
    y_pred_train = best_model.predict(X_train)
    train_score = f1_score(y_train, y_pred_train, average="macro")

    print(f"üèãÔ∏è F1-Score no Treino: {train_score:.4f}")

    # Gap treino-valida√ß√£o
    gap = train_score - grid_search.best_score_
    print(f"üìâ Gap Treino-Valida√ß√£o: {gap:.4f}")

    return best_model, grid_search
```

```python
def report_modelo(pipe, X_train, y_train, X_test, y_test):
    """
    Treina um pipeline de classifica√ß√£o, avalia o desempenho no conjunto de treino e teste,
    e imprime um relat√≥rio completo com m√©tricas e an√°lise de overfitting.

    M√©tricas inclu√≠das:
    - Acur√°cia (treino e teste)
    - F1-macro (treino e teste)
    - Matriz de confus√£o (teste)
    - Classification report detalhado (teste)
    - An√°lise de overfitting baseada na diferen√ßa de F1-macro entre treino e teste

    Par√¢metros:
    -----------
    pipe : sklearn Pipeline ou ImbPipeline
        Pipeline de modelagem contendo pr√©-processamento e estimador final.
    X_train : array-like ou DataFrame
        Features de treino.
    y_train : array-like ou Series
        Target de treino.
    X_test : array-like ou DataFrame
        Features de teste.
    y_test : array-like ou Series
        Target de teste.

    Observa√ß√µes:
    ------------
    - Funciona para classifica√ß√£o bin√°ria ou multiclasse.
    - Usa F1-macro para medir desempenho m√©dio por classe, sem pondera√ß√£o pelo tamanho das classes.
    - A an√°lise de overfitting √© categorizada como Baixo, Leve, Moderado ou Alto.
    """
    # Treinar
    pipe.fit(X_train, y_train)
    
    # Previs√µes
    y_pred_train = pipe.predict(X_train)
    y_pred_test = pipe.predict(X_test)
    
    # F1-macro
    f1_train = f1_score(y_train, y_pred_train, average='macro', zero_division=0)
    f1_test = f1_score(y_test, y_pred_test, average='macro', zero_division=0)
    
    # Acur√°cia geral
    acc_train = pipe.score(X_train, y_train)
    acc_test = pipe.score(X_test, y_test)
    
    # An√°lise de overfitting
    diff_f1 = f1_train - f1_test
    if diff_f1 < 0.05:
        overfitting = 'Baixo ou desprez√≠vel'
    elif diff_f1 < 0.10:
        overfitting = 'Leve'
    elif diff_f1 < 0.20:
        overfitting = 'Moderado'
    else:
        overfitting = 'Alto'
    
    # Matriz de confus√£o
    cm = confusion_matrix(y_test, y_pred_test)
    
    # Imprimindo resultados
    print("\n=== Report do Modelo ===\n")
    print(f"Acur√°cia Treino: {acc_train:.3f}")
    print(f"Acur√°cia Teste : {acc_test:.3f}")
    print(f"F1-macro Treino: {f1_train:.3f}")
    print(f"F1-macro Teste : {f1_test:.3f}")
    print(f"An√°lise de Overfitting: {overfitting}\n")
    
    print("--- Matriz de Confus√£o (Teste) ---")
    print(cm)
    
    print("\n--- Classification Report (Teste) ---")
    print(classification_report(y_test, y_pred_test, zero_division=0))
```

## Pr√©-processamento para os modelos

Como pr√©-processamento para o nosso modelo iremos realizar o tratamento das vari√°veis consideradas relevantes em nosso EDA, seguindo os seguintes passos:
1. Filtragem de colunas relevantes para o modelo

## 2. Otimiza√ß√£o - Random Forest (com SMOTE e StandardScaler)


Para o modelo de Random Forest foram escolhidos hiperpar√¢metros que controlam tanto a complexidade das √°rvores quanto a diversidade da floresta:

- n_estimators: n√∫mero de √°rvores na floresta; mais √°rvores podem aumentar a robustez do modelo, mas tamb√©m o custo computacional.

- max_depth: profundidade m√°xima das √°rvores; valores maiores aumentam a complexidade e podem levar a overfitting.

- min_samples_split: n√∫mero m√≠nimo de amostras necess√°rias para dividir um n√≥; valores mais altos resultam em √°rvores menos complexas.

- min_samples_leaf: n√∫mero m√≠nimo de amostras por folha; impede folhas muito pequenas, ajudando a reduzir overfitting.

- max_features: n√∫mero de vari√°veis consideradas em cada divis√£o; controla a aleatoriedade do modelo, influenciando a diversidade entre as √°rvores.

Esses hiperpar√¢metros foram otimizados para encontrar o melhor equil√≠brio entre desempenho.

```python
# =========================================================================
# OTIMIZA√á√ÉO RANDOM FOREST COM SMOTE E STANDARD SCALER
# =========================================================================

# Pipeline para Random Forest (com SMOTE e StandardScaler)
pipe_rf = ImbPipeline([
    ('scaler', StandardScaler()),
    ('smote', SMOTE(random_state=SEED)),
    ('model', RandomForestClassifier(random_state=SEED))
])

# Hiperpar√¢metros para otimiza√ß√£o do Random Forest
param_grid_rf = {
    'model__n_estimators': [50, 100, 200],
    'model__max_depth': [3, 5, 7],
    'model__min_samples_split': [2, 5, 10],
    'model__min_samples_leaf': [2, 4, 8],
    'model__max_features': ['sqrt', 'log2']
}

#best_rf_model, grid_search_rf = grid_optimization_parameters(
#    model="Random Forest", 
#    pipeline=pipe_rf, 
#    parameters=param_grid_rf, 
#    X_train=X_train, 
#    y_train=y_train
#)
```

O melhor modelo de random forest apresentou alguns √≠ndicios de overfitting com a performance caindo quase 0.07 pontos entre o treino e valida√ß√£o. Devido a isso ser√£o realizados alguns ajustes nos par√¢metros de otimiza√ß√£o

```python
# =========================================================================
# OTIMIZA√á√ÉO RANDOM FOREST COM SMOTE E STANDARD SCALER
# =========================================================================

# Pipeline para Random Forest (com SMOTE e StandardScaler)
pipe_rf = ImbPipeline([
    ('scaler', StandardScaler()),
    ('smote', SMOTE(random_state=SEED)),
    ('model', RandomForestClassifier(random_state=SEED))
])

# Hiperpar√¢metros para otimiza√ß√£o do Random Forest
param_grid_rf = {
    'model__n_estimators': [50, 100, 200],
    'model__max_depth': [3, 5, 7],
    'model__min_samples_split': [5, 10, 20], # Par√¢metro aumentado
    'model__min_samples_leaf': [4, 8, 16], # Par√¢metro aumentado
    'model__max_features': ['sqrt', 'log2']
}

#best_rf_model, grid_search_rf = grid_optimization_parameters(
#    model="Random Forest", 
#    pipeline=pipe_rf, 
#    parameters=param_grid_rf, 
#    X_train=X_train, 
#    y_train=y_train
#)
```

## 3. Otimiza√ß√£o - Logistic Regression (sem balanceamento e com StandardScaler)


Para o modelo de Logistic Regression foram escolhidos hiperpar√¢metros que controlam a regulariza√ß√£o e a complexidade do modelo:

- C: inverso da for√ßa da regulariza√ß√£o; valores menores aumentam a penaliza√ß√£o dos coeficientes, tornando o modelo mais simples e menos propenso a overfitting, enquanto valores maiores permitem maior flexibilidade.

- penalty: tipo de regulariza√ß√£o aplicada aos coeficientes; l1 pode zerar coeficientes e realizar sele√ß√£o de features, l2 suaviza os coeficientes, e elasticnet combina os dois m√©todos, equilibrando sele√ß√£o de features e regulariza√ß√£o.

- solver: algoritmo usado para otimiza√ß√£o da fun√ß√£o de custo; diferentes solvers suportam diferentes tipos de penalty, garantindo que o modelo seja trein√°vel de forma eficiente.

- l1_ratio: usado apenas com elasticnet; define a propor√ß√£o entre L1 e L2, permitindo ajustar o balanceamento entre sele√ß√£o de features e suaviza√ß√£o dos coeficientes.

Esses hiperpar√¢metros foram otimizados para encontrar o melhor equil√≠brio entre desempenho e capacidade de generaliza√ß√£o do modelo.

## Implementa√ß√£o do Modelo de Ensemble

Para criar um modelo de ensemble eficaz com os 3 modelos finais, utilizaremos o **VotingClassifier** do scikit-learn que implementa o majority voting. Isso significa que:

- Cada modelo faz sua predi√ß√£o individual
- A predi√ß√£o final √© determinada pela classe que recebe mais votos

Os 3 modelos que comp√µem o ensemble s√£o:
1. **Naive Bayes**
2. **Random Forest**
3. **Logistic Regression**


```python
# =========================================================================
# CRIA√á√ÉO DO VOTING CLASSIFIER (ENSEMBLE COM MAJORITY VOTING)
# =========================================================================

# Pipeline para Naive Bayes (sem balanceamento e escalonamento)
pipe_nb_ensemble = Pipeline([
    ('scaler', 'passthrough'),  # sem escalonamento
    ('model', GaussianNB(var_smoothing=1e-9))
])

# Pipeline para Random Forest (Upsampling, Standard Scaler)
pipe_rf_ensemble = ImbPipeline([
    ('scaler', StandardScaler()),
    ('smote', SMOTE(random_state=SEED)),
    ('model', RandomForestClassifier(
        max_depth=5, 
        max_features='sqrt',
        min_samples_leaf=4,
        min_samples_split=5,
        n_estimators=50,
        random_state=SEED))
])

# Pipeline para Logistic Regression (sem balanceamento e com StandardScaler)
pipe_lr_ensemble = Pipeline([
    ('scaler', StandardScaler()),
    ('model', LogisticRegression(max_iter=5000, C=0.001, penalty='l2', solver='liblinear', random_state=SEED))
])

# Criando o VotingClassifier com os 3 modelos
ensemble_model = VotingClassifier(
    estimators=[
        ('naive_bayes', pipe_nb_ensemble),
        ('random_forest', pipe_rf_ensemble), 
        ('logistic_regression', pipe_lr_ensemble)
    ],
    voting='hard'  # majority voting (hard voting)
)

# Treinando o modelo ensemble
ensemble_model.fit(X_train, y_train)
```

```python
# =========================================================================
# AVALIA√á√ÉO DO MODELO ENSEMBLE
# =========================================================================

print("üìä AVALIA√á√ÉO DO MODELO ENSEMBLE")
print("="*70)

# Fazendo predi√ß√µes com o ensemble
y_pred_ensemble_train = ensemble_model.predict(X_train)
y_pred_ensemble_test = ensemble_model.predict(X_test)

# Calculando m√©tricas do ensemble
ensemble_metrics = calculate_metrics(y_test, y_pred_ensemble_test, "Ensemble (Majority Voting)")

# Exibindo o report completo do ensemble
print("\nüéØ REPORT DETALHADO DO ENSEMBLE:")
report_modelo(ensemble_model, X_train, y_train, X_test, y_test)
```

Abaixo iremos fazer uma an√°lise comparat√≥rio entre os modelos 

```python
# =========================================================================
# COMPARA√á√ÉO: ENSEMBLE vs MODELOS INDIVIDUAIS
# =========================================================================

print("\nüìà COMPARA√á√ÉO DE PERFORMANCE: ENSEMBLE vs MODELOS INDIVIDUAIS")
print("="*80)

# Fazendo predi√ß√µes dos modelos individuais para compara√ß√£o
y_pred_nb = pipe_nb.predict(X_test)
y_pred_rf = pipe_rf.predict(X_test) 
y_pred_lr = pipe_lr.predict(X_test)

# Calculando m√©tricas de todos os modelos
metrics_comparison = []
metrics_comparison.append(calculate_metrics(y_test, y_pred_nb, "Naive Bayes"))
metrics_comparison.append(calculate_metrics(y_test, y_pred_rf, "Random Forest"))
metrics_comparison.append(calculate_metrics(y_test, y_pred_lr, "Logistic Regression"))
metrics_comparison.append(ensemble_metrics)

# Criando DataFrame para compara√ß√£o
df_comparison = pd.DataFrame(metrics_comparison)
df_comparison = df_comparison.round(3)

print("\nüìä TABELA COMPARATIVA DE M√âTRICAS:")
print("-"*80)
display(df_comparison)

```
