# <a id='descricao'>Descrição do Problema</a>

O conjunto de dados **HR Analytics** é um conjunto de dados multivariado que contém informações sobre funcionários de uma empresa, incluindo características demográficas, profissionais e comportamentais. O objetivo principal é analisar e prever a rotatividade de funcionários (attrition) com base em diversas variáveis como idade, salário, tempo de empresa, satisfação no trabalho, distância de casa, entre outras. O problema visa identificar fatores que contribuem para o desligamento de funcionários.

## Hipóteses do Problema

As hipóteses em relação ao **comportamento dos funcionários**:
- O nível de satisfação geral está baixo o que pode levar ao desligamento
- Os funcionários estão precisando viajar muito
- Os funcionários estão precisando fazer muitas horas extras
- Os funcionários estão recebendo poucos benefícios
- Os funcionários não estão conseguindo equilibrar a vida com o trabalho
- Os funcionários estão levando muito anos até serem promovidos
- Os funcionários estão levando muito tempo até chegar ao trabalho

As hipóteses que levam ao **desligamento da empresa**:
- Os funcionários mais novos estão se desligando mais da empresa por um comportamento da nova geração
- Pessoas que precisam de um deslocamento maior para ir trabalhar, também são mais propensas a deixar empresa
- Pessoas com salário menor são propensas a desligar da empresa
- Pessoas com mais tempo de empresa são menos propensas a deixar empresa
- Pessoas que realizam horas extras estão mais propensas a desligar da empresa

## Tipo de Problema

Este é um problema de **classificação supervisionada**. Dado um conjunto de características, o objetivo é prever se o funcionário irá deixar a empresa.

## Seleção de Dados

O dataset de Recursos humanos foi obtido no kaggle e contém diversas informações sobre funcionários de algumas empresas e se eles deixaram a empresa ou não. O dataset vai precisar de uma etapa de pré processamento para eliminar variáveis duplicadas como (Idade e grupo de idade) ou irrelevantes para análise como  "maior de 18 anos" (100% das instâncias são maiores de 18 anos).

### Tratamento de Warnings
- **Precisão/Recall indefinidos**: Alguns modelos podem não conseguir prever a classe minoritária, resultando em warnings. 
- **Solução implementada**: Uso de `make_scorer` com `zero_division=0` para tratar casos onde a precisão é indefinida.
- **Filtros de warnings**: Aplicados para manter o output limpo durante o grid search.

```python
def plot_comparative_boxplots(df_results, group_by='modelo', figsize=(14, 6)):
    """
    Plota dois boxplots comparativos na mesma figura: Acurácia e F1-Score
    
    Parâmetros:
    - df_results: DataFrame com os resultados dos experimentos
    - group_by: Variável para agrupar no eixo y ('balanceamento', 'scaler', 'modelo')
    - figsize: Tamanho da figura (largura, altura)
    """
    
    # Configuração da figura com 2 subplots
    fig, axes = plt.subplots(1, 2, figsize=figsize)
    
    # Configuração do estilo
    plt.style.use('default')
    
    # Boxplot da Acurácia
    sns.boxplot(data=df_results, y=group_by, x='acc_mean', hue=group_by, ax=axes[0], palette='viridis', legend=False)
    axes[0].set_title('Distribuição da Acurácia por ' + group_by.capitalize(), fontsize=14, fontweight='bold')
    axes[0].set_xlabel('Acurácia Média', fontsize=12)
    axes[0].set_ylabel(group_by.capitalize(), fontsize=12)
    axes[0].grid(True, alpha=0.3)
    
    # Boxplot do F1-Score
    sns.boxplot(data=df_results, y=group_by, x='f1_mean', hue=group_by, ax=axes[1], palette='plasma', legend=False)
    axes[1].set_title('Distribuição do F1-Score por ' + group_by.capitalize(), fontsize=14, fontweight='bold')
    axes[1].set_xlabel('F1-Score Médio', fontsize=12)
    axes[1].set_ylabel(group_by.capitalize(), fontsize=12)
    axes[1].grid(True, alpha=0.3)
    
    # Ajustar layout
    plt.tight_layout()
    
    # Mostrar o gráfico
    plt.show()
```

```python
def pipeline_combinacoes_modelos(X_train, y_train, models, balancing, scalers, n_splits=5, random_state=SEED):
    """
    Executa grid search sistemático testando todas as combinações de modelos, 
    técnicas de balanceamento e métodos de escalonamento.
    
    Parâmetros:
    -----------
    X_train : pd.DataFrame ou array-like
        Features de treino
    y_train : pd.Series ou array-like  
        Target de treino
    models : dict
        Dicionário com modelos {'nome': modelo_instancia}
    balancing : dict
        Dicionário com técnicas de balanceamento {'nome': tecnica_instancia}
    scalers : dict
        Dicionário com escalonadores {'nome': scaler_instancia}
    n_splits : int, default=5
        Número de folds para validação cruzada
    random_state : int, default=SEED
        Seed para reprodutibilidade
        
    Retorna:
    --------
    pd.DataFrame
        DataFrame com resultados ordenados por F1-Score decrescente, contendo:
        - balanceamento, scaler, modelo: combinação testada
        - acc_mean, acc_std: acurácia média e desvio padrão
        - f1_mean, f1_std: F1-Score macro médio e desvio padrão
        - precision_mean, precision_std: precisão macro média e desvio padrão
        - recall_mean, recall_std: recall macro médio e desvio padrão
        
    Notas:
    ------
    - Usa make_scorer com zero_division=0 para evitar warnings de precisão/recall indefinidos
    
    Exemplo:
    --------
    >>> resultados = pipeline_combinacoes_modelos(X_train, y_train, models, balancing, scalers)
    >>> print(resultados.head())
    """
    
    results = []
    cv = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=random_state)

    # Loop triplo para todas as combinações
    for bal_name, balancer in balancing.items():
        for scaler_name, scaler in scalers.items():
            for model_name, model in models.items():

                # Pipeline dinâmico baseado na presença de balanceamento
                if balancer:
                    pipe = ImbPipeline(steps=[
                        ("scaler", scaler),
                        ("bal", balancer),
                        ("model", model)
                    ])
                else:
                    pipe = Pipeline(steps=[
                        ("scaler", scaler),
                        ("model", model)
                    ])

                # Validação cruzada com múltiplas métricas
                acc_scores = cross_val_score(pipe, X_train, y_train, cv=cv, scoring="accuracy")
                f1_scores = cross_val_score(pipe, X_train, y_train, cv=cv, scoring="f1_macro")
                
                # Usando make_scorer para controlar o zero_division (evita warnings)
                precision_scorer = make_scorer(precision_score, average='macro', zero_division=0)
                recall_scorer = make_scorer(recall_score, average='macro', zero_division=0)
                
                precision_scores = cross_val_score(pipe, X_train, y_train, cv=cv, scoring=precision_scorer)
                recall_scores = cross_val_score(pipe, X_train, y_train, cv=cv, scoring=recall_scorer)

                results.append({
                    "balanceamento": bal_name,
                    "scaler": scaler_name,
                    "modelo": model_name,
                    "acc_mean": np.mean(acc_scores),
                    "acc_std": np.std(acc_scores),
                    "f1_mean": np.mean(f1_scores),
                    "f1_std": np.std(f1_scores),
                    "precision_mean": np.mean(precision_scores),
                    "precision_std": np.std(precision_scores),
                    "recall_mean": np.mean(recall_scores),
                    "recall_std": np.std(recall_scores)
                })

    # Retorna DataFrame ordenado por F1-Score
    df_results = pd.DataFrame(results)
    return df_results.sort_values(by="f1_mean", ascending=False)
    
```

## Pré-processamento para os modelos

Como pré-processamento para o nosso modelo iremos realizar o tratamento das variáveis consideradas relevantes em nosso EDA, seguindo os seguintes passos:
1. Filtragem de colunas relevantes para o modelo
2. Criação de dummies para variáveis categóricas e ajustes necessários
3. Separação em treino e teste (80/20)

```python
# Filtragem de colunas relevantes para o modelo
colunas_relevantes = ['Age', 'Attrition', 'JobLevel', 'JobRole', 'MonthlyIncome', 'OverTime', 'StockOptionLevel', 'YearsWithCurrManager']
df = df[colunas_relevantes]
```

Essas foram as colunas que mostraram melhor explicação do que leva um funcionário a sair da empresa.

```python
# Ajustes de colunas categóricas
## Mapeamento para variáveis binárias
df["Attrition"] = df["Attrition"].map({'Yes': 1, 'No': 0})
df["OverTime"] = df["OverTime"].map({'Yes': 1, 'No': 0})
df["FirstJobLevel"] = df["JobLevel"].map({1: 1, 2: 0, 3: 0, 4: 0, 5: 0})
df["StockOption"] = df["StockOptionLevel"].map({0: 0, 1: 1, 2: 1, 3: 1})


# Criação de dummies para variaveis categóricas
df = pd.get_dummies(df, columns=['JobRole'], prefix='JobRole')

# Ajustar nomes das colunas
df.columns = [col.replace("_", "").replace(" ", "") for col in df.columns]

# Ajustar o formato para inteiros
df['JobRoleHumanResources'] = df['JobRoleHumanResources'].astype(int)
df['JobRoleLaboratoryTechnician'] = df['JobRoleLaboratoryTechnician'].astype(int)
df['JobRoleSalesRepresentative'] = df['JobRoleSalesRepresentative'].astype(int)

# Dropar colunas que não serão utilizadas
df = df.drop(columns=[
    'JobLevel',
    'StockOptionLevel',
    'JobRoleHealthcareRepresentative',
    'JobRoleManager',
    'JobRoleManufacturingDirector',
    'JobRoleResearchDirector',
    'JobRoleResearchScientist',
    'JobRoleSalesExecutive'
 ])

df.head(3)
```

Ajuste das colunas para colunas numéricas para aplicação dos modelos

```python
# Separação em treino e teste (80/20)
X = df.drop('Attrition', axis=1)
y = df['Attrition']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=SEED, stratify=y)

print(f"X_train: {X_train.shape}")
print(f"X_test: {X_test.shape}")
print(f"y_train: {y_train.shape}")
print(f"y_test: {y_test.shape}")
```

# <a id='setup'>Comparativo de modelos e processamentos</a>

Nesta seção, iremos testar diferentes combinações de:

**Modelos**
- Regressão Logística
- Árvore de Decisão
- Random Forest 
- XGBoost Classifier
- Naive Bayes
- KNN
- SVM

**Escalonamento**
- Sem escalonamento
- MinMaxScaler
- StandardScaler

**Balanceamento**
- Sem balanceamento
- Upsampling (SMOTE)
- Downsampling

O objetivo é criar um pipeline sistemático que teste todas estas combinações e identifique o melhor modelo para cada tipo de balanceamento. Os modelos vencedores serão posteriormente otimizados através de ajuste de hiperparâmetros e avaliados no conjunto de teste.


```python
# === DEFINIÇÃO DE MODELOS ===
models = {
    "LogReg": LogisticRegression(max_iter=5000, random_state=SEED),
    "DecisionTree": DecisionTreeClassifier(random_state=SEED),
    "RandomForest": RandomForestClassifier(random_state=SEED),
    "XGBoost": XGBClassifier(eval_metric="logloss", random_state=SEED),
    "NaiveBayes": GaussianNB(),
    "KNN": KNeighborsClassifier(),
    "SVM": SVC(probability=True, random_state=SEED)
}

# === DEFINIÇÃO DE ESCALONADORES ===
scalers = {
    "none": "passthrough",  # sem transformação
    "standard": StandardScaler(),
    "minmax": MinMaxScaler()
}

# === DEFINIÇÃO DE ESTRATÉGIAS DE BALANCEAMENTO ===
balancing = {
    "none": None,
    "smote": SMOTE(random_state=SEED),
    "undersample": RandomUnderSampler(random_state=SEED)
}
```

```python
# =========================================================================
# EXECUÇÃO DO GRID SEARCH COMPLETO COM MÚLTIPLAS MÉTRICAS
# =========================================================================
"""
Executando teste sistemático de todas as combinações:
📊 3 Balanceamentos × 3 Scalers × 7 Modelos = 63 combinações
📈 Métricas avaliadas: Acurácia, F1-Score, Precisão e Recall (todas macro)
🔄 Validação cruzada estratificada com 5 folds
"""

df_results_1 = pipeline_combinacoes_modelos(
    X_train=X_train, 
    y_train=y_train, 
    models=models, 
    balancing=balancing, 
    scalers=scalers,
    n_splits=5,
    random_state=SEED
)
```

```python
# Visualizações comparativas
plot_comparative_boxplots(df_results_1, group_by='balanceamento')
plot_comparative_boxplots(df_results_1, group_by='scaler')
plot_comparative_boxplots(df_results_1, group_by='modelo')

# Rankings por diferentes métricas
print("🏆 TOP 3 MELHORES COMBINAÇÕES POR MÉTRICA")
print("="*50)

print("\n📊 Melhores por F1-Score:")
display(df_results_1.sort_values(by='f1_mean', ascending=False).head(3))

print("\n✅ Melhores por Acurácia:")
display(df_results_1.sort_values(by='acc_mean', ascending=False).head(3))
```
