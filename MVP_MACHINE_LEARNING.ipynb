# <a id='descricao'>Descrição do Problema</a>

O conjunto de dados **HR Analytics** é um conjunto de dados multivariado que contém informações sobre funcionários de uma empresa, incluindo características demográficas, profissionais e comportamentais. O objetivo principal é analisar e prever a rotatividade de funcionários (attrition) com base em diversas variáveis como idade, salário, tempo de empresa, satisfação no trabalho, distância de casa, entre outras. O problema visa identificar fatores que contribuem para o desligamento de funcionários.

## Hipóteses do Problema

As hipóteses em relação ao **comportamento dos funcionários**:
- O nível de satisfação geral está baixo o que pode levar ao desligamento
- Os funcionários estão precisando viajar muito
- Os funcionários estão precisando fazer muitas horas extras
- Os funcionários estão recebendo poucos benefícios
- Os funcionários não estão conseguindo equilibrar a vida com o trabalho
- Os funcionários estão levando muito anos até serem promovidos
- Os funcionários estão levando muito tempo até chegar ao trabalho

As hipóteses que levam ao **desligamento da empresa**:
- Os funcionários mais novos estão se desligando mais da empresa por um comportamento da nova geração
- Pessoas que precisam de um deslocamento maior para ir trabalhar, também são mais propensas a deixar empresa
- Pessoas com salário menor são propensas a desligar da empresa
- Pessoas com mais tempo de empresa são menos propensas a deixar empresa
- Pessoas que realizam horas extras estão mais propensas a desligar da empresa

## Tipo de Problema

Este é um problema de **classificação supervisionada**. Dado um conjunto de características, o objetivo é prever se o funcionário irá deixar a empresa.

## Seleção de Dados

O dataset de Recursos humanos foi obtido no kaggle e contém diversas informações sobre funcionários de algumas empresas e se eles deixaram a empresa ou não. O dataset vai precisar de uma etapa de pré processamento para eliminar variáveis duplicadas como (Idade e grupo de idade) ou irrelevantes para análise como  "maior de 18 anos" (100% das instâncias são maiores de 18 anos).

```python
def plot_comparative_boxplots(df_results, group_by='modelo', figsize=(14, 6)):
    """
    Plota dois boxplots comparativos na mesma figura: Acurácia e F1-Score
    
    Parâmetros:
    - df_results: DataFrame com os resultados dos experimentos
    - group_by: Variável para agrupar no eixo y ('balanceamento', 'scaler', 'modelo')
    - figsize: Tamanho da figura (largura, altura)
    """
    
    # Configuração da figura com 2 subplots
    fig, axes = plt.subplots(1, 2, figsize=figsize)
    
    # Configuração do estilo
    plt.style.use('default')
    
    # Boxplot da Acurácia
    sns.boxplot(data=df_results, y=group_by, x='acc_mean', hue=group_by, ax=axes[0], palette='viridis', legend=False)
    axes[0].set_title('Distribuição da Acurácia por ' + group_by.capitalize(), fontsize=14, fontweight='bold')
    axes[0].set_xlabel('Acurácia Média', fontsize=12)
    axes[0].set_ylabel(group_by.capitalize(), fontsize=12)
    
    # Boxplot do F1-Score
    sns.boxplot(data=df_results, y=group_by, x='f1_mean', hue=group_by, ax=axes[1], palette='plasma', legend=False)
    axes[1].set_title('Distribuição do F1-Score por ' + group_by.capitalize(), fontsize=14, fontweight='bold')
    axes[1].set_xlabel('F1-Score Médio', fontsize=12)
    axes[1].set_ylabel(group_by.capitalize(), fontsize=12)
    
    # Ajustar layout
    plt.tight_layout()
    
    # Mostrar o gráfico
    plt.show()
```

```python
def pipeline_models_combination(X_train, y_train, models, balancing, scalers, n_splits=5, random_state=SEED):
    """
    Executa grid search sistemático testando todas as combinações de modelos, 
    técnicas de balanceamento e métodos de escalonamento.
    
    Parâmetros:
    -----------
    X_train : pd.DataFrame ou array-like
        Features de treino
    y_train : pd.Series ou array-like  
        Target de treino
    models : dict
        Dicionário com modelos {'nome': modelo_instancia}
    balancing : dict
        Dicionário com técnicas de balanceamento {'nome': tecnica_instancia}
    scalers : dict
        Dicionário com escalonadores {'nome': scaler_instancia}
    n_splits : int, default=5
        Número de folds para validação cruzada
    random_state : int, default=SEED
        Seed para reprodutibilidade
        
    Retorna:
    --------
    pd.DataFrame
        DataFrame com resultados ordenados por F1-Score decrescente, contendo:
        - balanceamento, scaler, modelo: combinação testada
        - acc_mean, acc_std: acurácia média e desvio padrão
        - f1_mean, f1_std: F1-Score macro médio e desvio padrão
        - precision_mean, precision_std: precisão macro média e desvio padrão
        - recall_mean, recall_std: recall macro médio e desvio padrão
        
    Notas:
    ------
    - Usa make_scorer com zero_division=0 para evitar warnings de precisão/recall indefinidos
    
    Exemplo:
    --------
    >>> resultados = pipeline_combinacoes_modelos(X_train, y_train, models, balancing, scalers)
    >>> print(resultados.head())
    """
    
    results = []
    cv = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=random_state)

    # Loop triplo para todas as combinações
    for bal_name, balancer in balancing.items():
        for scaler_name, scaler in scalers.items():
            for model_name, model in models.items():

                # Pipeline dinâmico baseado na presença de balanceamento
                if balancer:
                    pipe = ImbPipeline(steps=[
                        ("scaler", scaler),
                        ("bal", balancer),
                        ("model", model)
                    ])
                else:
                    pipe = Pipeline(steps=[
                        ("scaler", scaler),
                        ("model", model)
                    ])

                # Validação cruzada com múltiplas métricas
                acc_scores = cross_val_score(pipe, X_train, y_train, cv=cv, scoring="accuracy")
                f1_scores = cross_val_score(pipe, X_train, y_train, cv=cv, scoring="f1_macro")
                
                # Usando make_scorer para controlar o zero_division (evita warnings)
                precision_scorer = make_scorer(precision_score, average='macro', zero_division=0)
                recall_scorer = make_scorer(recall_score, average='macro', zero_division=0)
                
                precision_scores = cross_val_score(pipe, X_train, y_train, cv=cv, scoring=precision_scorer)
                recall_scores = cross_val_score(pipe, X_train, y_train, cv=cv, scoring=recall_scorer)

                results.append({
                    "balanceamento": bal_name,
                    "scaler": scaler_name,
                    "modelo": model_name,
                    "acc_mean": np.mean(acc_scores),
                    "acc_std": np.std(acc_scores),
                    "f1_mean": np.mean(f1_scores),
                    "f1_std": np.std(f1_scores),
                    "precision_mean": np.mean(precision_scores),
                    "precision_std": np.std(precision_scores),
                    "recall_mean": np.mean(recall_scores),
                    "recall_std": np.std(recall_scores)
                })

    # Retorna DataFrame ordenado por F1-Score
    df_results = pd.DataFrame(results)
    return df_results.sort_values(by="f1_mean", ascending=False)
    
```

```python
def calculate_metrics(y_true, y_pred, model_name):
    """
    Calcula métricas de classificação (macro) e retorna em formato de dicionário.

    Parâmetros
    ----------
    y_true : array-like
        Rótulos verdadeiros.
    y_pred : array-like
        Rótulos preditos.
    model_name : str
        Nome do modelo avaliado.

    Retorna
    -------
    dict
        Dicionário com: nome do modelo, acurácia, F1-score, precisão e recall (macro).
    """
    return {
        'Modelo': model_name,
        'Acurácia': accuracy_score(y_true, y_pred),
        'F1-Score': f1_score(y_true, y_pred, average='macro'),
        'Precisão': precision_score(y_true, y_pred, average='macro', zero_division=0),
        'Recall': recall_score(y_true, y_pred, average='macro', zero_division=0)
    }
```

```python
def grid_optimization_parameters(model, pipeline, parameters, X_train, y_train, random_state=SEED):
    """
    Executa a otimização de hiperparâmetros utilizando GridSearchCV com validação cruzada estratificada,
    avalia o melhor modelo encontrado e verifica possíveis sinais de overfitting.

    Parâmetros
    ----------
    model : str
        Nome do modelo (usado apenas para logging/prints).
    pipeline : sklearn.pipeline.Pipeline
        Pipeline de pré-processamento + modelo a ser otimizado.
    parameters : dict
        Grade de hiperparâmetros a ser explorada pelo GridSearchCV.
    X_train : array-like ou pandas.DataFrame
        Conjunto de features de treino.
    y_train : array-like ou pandas.Series
        Conjunto de rótulos de treino.
    random_state : int
        Semente para reprodutibilidade na validação cruzada.

    Retorna
    -------
    best_model : sklearn estimator
        Estimador treinado com os melhores hiperparâmetros encontrados.
    grid_search : GridSearchCV
        Objeto do GridSearchCV com todos os resultados da busca.

    Notas
    -----
    - A função imprime os melhores hiperparâmetros, o F1-score médio de validação 
      (via cross-validation) e o F1-score no treino com os melhores parâmetros.
    - Também calcula o "gap" entre treino e validação para indicar possíveis sinais de overfitting.
    - A métrica utilizada na otimização é o F1-macro, mas o cálculo do gap pode ser adaptado 
      para outras métricas, se necessário.
    """
    print(f"📊 Iniciando otimização de hiperparâmetros - {model}")
    print("=" * 60)

    # GridSearchCV
    grid_search = GridSearchCV(
        pipeline,
        parameters,
        cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=random_state),
        scoring='f1_macro',
        n_jobs=-1,
        verbose=1
    )

    # Treinamento
    grid_search.fit(X_train, y_train)

    # Resultados
    print(f"\n✅ Melhores hiperparâmetros {model}: {grid_search.best_params_}")
    print(f"🎯 Melhor F1-Score (validação - crossval): {grid_search.best_score_:.4f}")

    # Modelo otimizado
    best_model = grid_search.best_estimator_

    # Score treino com os melhores parâmetros
    y_pred_train = best_model.predict(X_train)
    train_score = f1_score(y_train, y_pred_train, average="macro")

    print(f"🏋️ F1-Score no Treino: {train_score:.4f}")

    # Gap treino-validação
    gap = train_score - grid_search.best_score_
    print(f"📉 Gap Treino-Validação: {gap:.4f}")

    return best_model, grid_search
```

```python
def report_modelo(pipe, X_train, y_train, X_test, y_test):
    """
    Treina um pipeline de classificação, avalia o desempenho no conjunto de treino e teste,
    e imprime um relatório completo com métricas e análise de overfitting.

    Métricas incluídas:
    - Acurácia (treino e teste)
    - F1-macro (treino e teste)
    - Matriz de confusão (teste)
    - Classification report detalhado (teste)
    - Análise de overfitting baseada na diferença de F1-macro entre treino e teste

    Parâmetros:
    -----------
    pipe : sklearn Pipeline ou ImbPipeline
        Pipeline de modelagem contendo pré-processamento e estimador final.
    X_train : array-like ou DataFrame
        Features de treino.
    y_train : array-like ou Series
        Target de treino.
    X_test : array-like ou DataFrame
        Features de teste.
    y_test : array-like ou Series
        Target de teste.

    Observações:
    ------------
    - Funciona para classificação binária ou multiclasse.
    - Usa F1-macro para medir desempenho médio por classe, sem ponderação pelo tamanho das classes.
    - A análise de overfitting é categorizada como Baixo, Leve, Moderado ou Alto.
    """
    # Treinar
    pipe.fit(X_train, y_train)
    
    # Previsões
    y_pred_train = pipe.predict(X_train)
    y_pred_test = pipe.predict(X_test)
    
    # F1-macro
    f1_train = f1_score(y_train, y_pred_train, average='macro', zero_division=0)
    f1_test = f1_score(y_test, y_pred_test, average='macro', zero_division=0)
    
    # Acurácia geral
    acc_train = pipe.score(X_train, y_train)
    acc_test = pipe.score(X_test, y_test)
    
    # Análise de overfitting
    diff_f1 = f1_train - f1_test
    if diff_f1 < 0.05:
        overfitting = 'Baixo ou desprezível'
    elif diff_f1 < 0.10:
        overfitting = 'Leve'
    elif diff_f1 < 0.20:
        overfitting = 'Moderado'
    else:
        overfitting = 'Alto'
    
    # Matriz de confusão
    cm = confusion_matrix(y_test, y_pred_test)
    
    # Imprimindo resultados
    print("\n=== Report do Modelo ===\n")
    print(f"Acurácia Treino: {acc_train:.3f}")
    print(f"Acurácia Teste : {acc_test:.3f}")
    print(f"F1-macro Treino: {f1_train:.3f}")
    print(f"F1-macro Teste : {f1_test:.3f}")
    print(f"Análise de Overfitting: {overfitting}\n")
    
    print("--- Matriz de Confusão (Teste) ---")
    print(cm)
    
    print("\n--- Classification Report (Teste) ---")
    print(classification_report(y_test, y_pred_test, zero_division=0))
```

# <a id='ensemble'>Agregação dos modelos</a>

## Implementação do Modelo de Ensemble

Para criar um modelo de ensemble eficaz com os 3 modelos finais, utilizaremos o **VotingClassifier** do scikit-learn que implementa o majority voting. Isso significa que:

- Cada modelo faz sua predição individual
- A predição final é determinada pela classe que recebe mais votos

Os 3 modelos que compõem o ensemble são:
1. **Naive Bayes**
2. **Random Forest**
3. **Logistic Regression**


```python
# =========================================================================
# CRIAÇÃO DO VOTING CLASSIFIER (ENSEMBLE COM MAJORITY VOTING)
# =========================================================================

# Pipeline para Naive Bayes (sem balanceamento e escalonamento)
pipe_nb_ensemble = Pipeline([
    ('scaler', 'passthrough'),  # sem escalonamento
    ('model', GaussianNB(var_smoothing=1e-9))
])

# Pipeline para Random Forest (Upsampling, Standard Scaler)
pipe_rf_ensemble = ImbPipeline([
    ('scaler', StandardScaler()),
    ('smote', SMOTE(random_state=SEED)),
    ('model', RandomForestClassifier(
        max_depth=5, 
        max_features='sqrt',
        min_samples_leaf=4,
        min_samples_split=5,
        n_estimators=50,
        random_state=SEED))
])

# Pipeline para Logistic Regression (sem balanceamento e com StandardScaler)
pipe_lr_ensemble = Pipeline([
    ('scaler', StandardScaler()),
    ('model', LogisticRegression(max_iter=5000, C=0.001, penalty='l2', solver='liblinear', random_state=SEED))
])

# Criando o VotingClassifier com os 3 modelos
ensemble_model = VotingClassifier(
    estimators=[
        ('naive_bayes', pipe_nb_ensemble),
        ('random_forest', pipe_rf_ensemble), 
        ('logistic_regression', pipe_lr_ensemble)
    ],
    voting='hard'  # majority voting (hard voting)
)

# Treinando o modelo ensemble
ensemble_model.fit(X_train, y_train)
```

```python
# =========================================================================
# AVALIAÇÃO DO MODELO ENSEMBLE
# =========================================================================

print("📊 AVALIAÇÃO DO MODELO ENSEMBLE")
print("="*70)

# Fazendo predições com o ensemble
y_pred_ensemble_train = ensemble_model.predict(X_train)
y_pred_ensemble_test = ensemble_model.predict(X_test)

# Calculando métricas do ensemble
ensemble_metrics = calculate_metrics(y_test, y_pred_ensemble_test, "Ensemble (Majority Voting)")

# Exibindo o report completo do ensemble
print("\n🎯 REPORT DETALHADO DO ENSEMBLE:")
report_modelo(ensemble_model, X_train, y_train, X_test, y_test)
```

O modelo de ensemble teve um comportamento bem semelhante aos outros modelos, com boas métricas para a classe de funcionários que permaneceram na empresa e problemas de performance na identificação de funcionários que vão deixar a empresa. Abaixo iremos fazer uma análise comparatório entre os modelos 

```python
# =========================================================================
# COMPARAÇÃO: ENSEMBLE vs MODELOS INDIVIDUAIS
# =========================================================================

print("\n📈 COMPARAÇÃO DE PERFORMANCE: ENSEMBLE vs MODELOS INDIVIDUAIS")
print("="*80)

# Fazendo predições dos modelos individuais para comparação
y_pred_nb = pipe_nb.predict(X_test)
y_pred_rf = pipe_rf.predict(X_test) 
y_pred_lr = pipe_lr.predict(X_test)

# Calculando métricas de todos os modelos
metrics_comparison = []
metrics_comparison.append(calculate_metrics(y_test, y_pred_nb, "Naive Bayes"))
metrics_comparison.append(calculate_metrics(y_test, y_pred_rf, "Random Forest"))
metrics_comparison.append(calculate_metrics(y_test, y_pred_lr, "Logistic Regression"))
metrics_comparison.append(ensemble_metrics)

# Criando DataFrame para comparação
df_comparison = pd.DataFrame(metrics_comparison)
df_comparison = df_comparison.round(3)

print("\n📊 TABELA COMPARATIVA DE MÉTRICAS:")
print("-"*80)
display(df_comparison)

```
